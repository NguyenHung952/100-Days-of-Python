#**📌 Bài 1 – Tính số năm còn lại đến khi nghỉ hưu**

Viết một chương trình:

Hỏi tuổi người dùng

Tính số năm còn lại đến 60 tuổi

In ra: “Bạn còn X năm nữa là đến tuổi nghỉ hưu.”

```python
# Gợi ý đoạn code khởi đầu
age = int(input("Nhập tuổi hiện tại của bạn: "))
years_left = 60 - age
print(f"Bạn còn {years_left} năm nữa là đến tuổi nghỉ hưu.")
```

#**📌 Bài 2 – Chuyển đổi tiền tệ đơn giản**
Hỏi người dùng muốn đổi bao nhiêu đô la Mỹ (USD)

Dùng tỉ giá cố định (VD: 1 USD = 24,000 VND)

In ra số tiền quy đổi được sang VND

```python
usd = float(input("Nhập số USD bạn muốn đổi: "))
vnd = usd * 24000
print(f"Số tiền sau khi quy đổi: {vnd:,.0f} VND")
```
