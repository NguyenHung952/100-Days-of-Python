#**✅ Bài 1 – Máy tính chỉ số BMI**
Nhập chiều cao (m) và cân nặng (kg)

Tính BMI theo công thức: BMI = weight / (height * height)

In ra chỉ số BMI và phân loại (nếu muốn nâng cao)

```python
height = float(input("Nhập chiều cao (m):"))
weight = float(input("Nhập cân nặng (kg):"))
BM = weight / (height * height)
print ("Chỉ số BMI của bạn là:", BM)
```

#**✅ Bài 2 – Phân tích chuỗi số điện thoại**
Nhập chuỗi số điện thoại (VD: "0912345678")

In ra:

Mã vùng (3 số đầu)

Số thuê bao (7 số sau)

Tổng độ dài chuỗi

```python
sdt=input("Nhập chuỗi số điện thoại:")
print("Mã vùng:",sdt[0:3])
print("Số thuê bao:",sdt[3:10])
print("Tổng độ dài chuỗi:",len(sdt))
```

#**✅ Bài 3 – Bộ đếm ký tự trong chuỗi**
Nhập một chuỗi bất kỳ từ người dùng

In ra độ dài chuỗi và số lần xuất hiện của một ký tự cụ thể (VD: chữ “a”)

```python
chuoi=input("Nhập chuỗi bất kỳ:")
print("Độ dài chuỗi:",len(chuoi))
print("Số lần xuất hiện của ký tự a:",chuoi.count("a"))
```
