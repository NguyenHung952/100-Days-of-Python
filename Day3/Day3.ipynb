#**✅ Bài 1 – Máy tính tuổi đủ điều kiện thi bằng lái**
Nhập tuổi người dùng

Nếu ≥18 tuổi → in: “Bạn đủ điều kiện thi bằng lái xe.”

Ngược lại → in: “Bạn chưa đủ tuổi.”

```python
tuoi=int(input("Nhập tuổi: "))
if(tuoi>=18):
  print("Bạn đủ điều kiện thi bằng lái xe.")
else:
  print("Bạn chưa đủ tuổi.")
```

#**✅ Bài 2 – Kiểm tra năm nhuận**
Nhập một năm bất kỳ

Kiểm tra xem có phải là năm nhuận hay không theo quy tắc:

Chia hết cho 4 và không chia hết cho 100

Hoặc chia hết cho 400

```python
nam = int(input("Nhập năm: "))
if (nam % 4 == 0 and nam % 100 != 0) or (nam % 400 == 0):
  print(nam, "là năm nhuận")
else:
  print(nam, "không phải là năm nhuận")
```

#**✅ Bài 3 – Máy tính hóa đơn tiền điện**
Nhập số kWh tiêu thụ

Tính tiền điện theo bảng giá đơn giản (VD):

≤ 50 kWh: 1.500 đ/kWh

51–100 kWh: 1.800 đ/kWh

100 kWh: 2.000 đ/kWh

```python
dien=int(input("Nhập số kWh tiêu thụ: "))
if(dien<=50):
  print("Tiền điện là: ", dien*1500)
if(dien>50 and dien<=100):
  print("Tiền điện là: ", dien*1800)
if(dien>100):
  print("Tiền điện là: ", dien*2000)
```
