#✅ Bài 1 – Tính tổng các số từ 1 đến n
Nhập số nguyên dương n từ người dùng

In ra tổng từ 1 đến n

```python
n=int(input("Nhập số nguyên dương n: "))
tong=0
for i in range(1,n+1):
    tong+=i
print("Tổng từ 1 đến",n,"là:",tong)
```

#✅ Bài 2 – In ra số chẵn từ 1 đến 100
Duyệt các số từ 1 đến 100

In ra các số chẵn

#✅ Bài 3 – Máy tính điểm trung bình
Nhập danh sách điểm (ngăn cách bằng dấu phẩy)

Tính điểm trung bình

In ra kết quả và đánh giá: giỏi (>=8), khá (>=6.5), trung bình (<6.5)

```python
danhsach=input("Nhập danh sách điểm (ngăn cách bằng dấu phẩy): ")
diem=danhsach.split(",")
tong=0
for i in diem:
    tong+=float(i)
tb=tong/len(diem)
if tb>=8:
    print("Điểm trung bình là:",tb,"Giỏi")
elif tb>=6.5:
    print("Điểm trung bình là:",tb,"Khá")
else:
    print("Điểm trung bình là:",tb,"Trung bình")
```

#✅ Bài 4 – Trò chơi đoán số (Mini game)
Tạo một số bí mật ngẫu nhiên từ 1 đến 100

Cho người chơi đoán đến khi đúng

Mỗi lần đoán, in ra: "Lớn hơn", "Nhỏ hơn", hoặc "Đúng rồi!"

```python
import random
n=random.randint(1,100)
a=input("Nhấn enter để bắt đầu trò chơi")
while True:
    try:
        guess = int(input("Hãy đoán một số từ 1 đến 100: "))
        if guess < n:
            print("Lớn hơn")
        elif guess > n:
            print("Nhỏ hơn")
        else:
            print("Đúng rồi!")
            break
    except ValueError:
        print("Vui lòng nhập một số nguyên hợp lệ.")
```
