#**✅ Bài 1 – Tung đồng xu mô phỏng**
Mô phỏng trò tung đồng xu

In ra kết quả: “Mặt ngửa” hoặc “Mặt sấp” (random)

```python
import random
dongxu=["Mặt ngửa", "Mặt sấp"]
print(random.choice(dongxu))
```

#**✅ Bài 2 – Người nào trả tiền? (Who Pays)**
Nhập danh sách tên (ngăn cách bằng dấu phẩy)

Dùng random.choice() chọn ngẫu nhiên 1 người → In kết quả: “Hôm nay, [Tên] sẽ trả tiền!”

```python
import random
names_string = input("Nhập danh sách tên (ngăn cách bằng dấu phẩy): ")
names = names_string.split(", ")
print(f"Hôm nay, {random.choice(names)} sẽ trả tiền!")
```

#**✅ Bài 3 – Bàn cờ kho báu (Treasure Map)**
Tạo bản đồ 3x3 dùng danh sách lồng nhau (list of lists)

Cho người dùng nhập vị trí (VD: "23" là dòng 2, cột 3)

Đặt chữ "X" vào vị trí đó


# Task
Hướng dẫn cách cập nhật bản đồ 3x3 bằng cách nhập vị trí từ người dùng và đặt ký tự "X" vào vị trí đó.

## Tạo bản đồ

### Subtask:
Tạo một danh sách lồng nhau (list of lists) đại diện cho bản đồ 3x3.


**Reasoning**:
Create a 3x3 list of lists representing the map and print it.



```python
map = [[" ", " ", " "], [" ", " ", " "], [" ", " ", " "]]
print(map)
```

## Nhận input từ người dùng

### Subtask:
Yêu cầu người dùng nhập vị trí dưới dạng chuỗi (ví dụ: "23").


**Reasoning**:
Use the input function to get the position from the user and store it in a variable.



```python
position = input("Nhập vị trí bạn muốn đặt kho báu (ví dụ: 23): ")
```

## Xử lý input

### Subtask:
Phân tích chuỗi input để lấy ra số dòng và số cột.


**Reasoning**:
Extract the column and row indices from the input string and convert them to zero-based integers.



```python
column_index = int(position[0]) - 1
row_index = int(position[1]) - 1
```

## Cập nhật bản đồ

### Subtask:
Đặt ký tự "X" vào vị trí được chỉ định trên bản đồ.


**Reasoning**:
Update the map by placing 'X' at the calculated row and column indices and then print the updated map.



```python
map[row_index][column_index] = "X"
print(map)
```

## Hiển thị bản đồ

### Subtask:
In bản đồ ra màn hình để người dùng xem kết quả.


**Reasoning**:
Print the `map` variable to display the current state of the map after placing the "X".



```python
display(map)
```
